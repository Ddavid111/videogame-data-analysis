```python
import warnings
import logging
import importlib
import config as split_config
importlib.reload(split_config)
MERGED_MASTER_FILE = split_config.MERGED_MASTER_FILE
MEDIA_TABLES_PATH = split_config.MEDIA_TABLES_PATH
SUPPORT_TABLES_PATH = split_config.SUPPORT_TABLES_PATH
REQUIREMENTS_TABLES_PATH = split_config.REQUIREMENTS_TABLES_PATH
PLATFORMS_TABLES_PATH = split_config.PLATFORMS_TABLES_PATH
PACKAGES_TABLES_PATH = split_config.PACKAGES_TABLES_PATH
PEOPLE_TABLES_PATH = split_config.PEOPLE_TABLES_PATH
GENRES_TABLES_PATH = split_config.GENRES_TABLES_PATH
TAGS_TABLES_PATH = split_config.TAGS_TABLES_PATH
METADATA_TABLES_PATH = split_config.METADATA_TABLES_PATH
LANGUAGE_TABLES_PATH = split_config.LANGUAGE_TABLES_PATH
from utils.io_utils import load_csv_safely

# ======== MODULIMPORTOK ========
from tables.media_tables import (
    create_media_table,
    create_screenshots_table,
    create_movies_table,
)
from tables.support_tables import create_support_table
from tables.requirements_tables import create_requirements_table
from tables.platforms_tables import create_platforms_table
from tables.package_tables import create_packages_table
from tables.developer_publisher_tables import (
    create_developer_tables,
    create_publisher_tables,
)
from tables.genre_category_tables import (
    create_genres_table,
    create_categories_table,
)
from tables.tags_tables import create_tags_table
from tables.game_metadata import (
    create_description_table,
    create_game_table,
)
from tables.language_tables import create_languages_table


# ======== SPLIT PIPELINE ========
def main():
    warnings.filterwarnings("ignore", category=FutureWarning)
    logging.info("=== Starting splitting process ===")

    # --- BetĂ¶ltĂ©s ---
    D = load_csv_safely(MERGED_MASTER_FILE)
    if D.empty:
        logging.error(f"Failed to load data from {MERGED_MASTER_FILE}")
        return

    logging.info(f"Loaded merged master with {len(D)} rows and {len(D.columns)} columns.")
    logging.info("========================================")

    table_funcs = [
        ("Media", create_media_table, MEDIA_TABLES_PATH),
        ("Screenshots", create_screenshots_table, MEDIA_TABLES_PATH),
        ("Movies", create_movies_table, MEDIA_TABLES_PATH),
        ("Support", create_support_table, SUPPORT_TABLES_PATH),
        ("Requirements", create_requirements_table, REQUIREMENTS_TABLES_PATH),
        ("Platforms", create_platforms_table, PLATFORMS_TABLES_PATH),
        ("Packages", create_packages_table, PACKAGES_TABLES_PATH),
        ("Developers", create_developer_tables, PEOPLE_TABLES_PATH),
        ("Publishers", create_publisher_tables, PEOPLE_TABLES_PATH),
        ("Genres", create_genres_table, GENRES_TABLES_PATH),
        ("Categories", create_categories_table, GENRES_TABLES_PATH),
        ("Tags", create_tags_table, TAGS_TABLES_PATH),
        ("Description", create_description_table, METADATA_TABLES_PATH),
        ("Game", create_game_table, METADATA_TABLES_PATH),
        ("Languages", create_languages_table, LANGUAGE_TABLES_PATH),
    ]

    for name, func, out_dir in table_funcs:
        try:
            logging.info(f"--- Generating {name} table ---")
            func(D, output_dir=out_dir)
        except Exception as e:
            logging.error(f"Error generating {name} table: {e}")

    logging.info("=== Splitting process completed successfully! ===")


if __name__ == "__main__":
    main()
```
